```python
import pandas as pd
database=pd.read_csv('./FinalDataBase.csv')
database
database["Start Date"] = pd.to_datetime(database["Start Date"], format="%d-%m-%Y")
database["End Date"] = pd.to_datetime(database["End Date"], format="%d-%m-%Y")
#Getting The Database Which has The Mapping Of Symbol-Company Name-Accord Code
```

```python
stock_price=pd.read_csv('./All Close Prices.csv')
stock_price['Date'] = pd.to_datetime(stock_price['Date']).dt.strftime('%d-%m-%Y')
stock_price=stock_price[stock_price['Exchange']=='NSE']

stock_price=stock_price.sort_values(by='Date',ascending=True)
#Getting The Stock Data And Only Getting The NSE Stock
```

```python
stock_list=database['Stock'].unique().tolist()
```

```python
from datetime import datetime
months=[1,7]
global_arr=[]
for year in range(2013,2023):
    for month in months:
        for stock in stock_list:
            company_name=database[database['Stock']==stock].iloc[0]['Company Name']
            current_date=datetime(year,month,1).date()
            stock_data=database[database['Stock']==stock]
            tell=False
            for row in stock_data.iterrows():
                row_data=row[1]
                if (row_data['Start Date'].date()<=current_date) and (current_date<row_data['End Date'].date()):
                    tell=True
            if tell:
                global_arr.append({
                    'Date':current_date,
                    'Symbol':stock
                })
            
             
```
